```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("Disease_with_Symptoms.csv")
df.head(20)
```

```python
df.shape
```

```python
#Since the values of the features are in the categorical format we have to do one hot encoding
data=pd.get_dummies(df.filter(regex='Symptom'),prefix='',prefix_sep='')
```

```python
data.head()
```

```python
data=data.groupby(data.columns,axis=1).agg(np.max)
```

```python
data.sample(20)
```

```python
filtered_data=data.iloc[:,1:-4]
```

```python
filtered_data.sample(10)
```

```python
filtered_data.shape
```

```python
Symptoms=filtered_data.columns
```

```python
filtered_data=pd.merge(filtered_data,df[['Code','Disease']],left_index=True,right_index=True)
```

```python
disease_list=filtered_data['Disease'].explode().tolist()
```

```python
disease_list
```

```python
len(disease_list)
```

```python
from collections import OrderedDict
disease = list(OrderedDict.fromkeys(disease_list))
```

```python
disease
```

```python
X=filtered_data.iloc[:,:-2]
Y=filtered_data.iloc[:,-2]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,random_state=5,test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model=RandomForestClassifier(n_estimators=40)
model.fit(X_train,Y_train)
```

```python
Y_predict=model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy=accuracy_score(Y_test,Y_predict)*100
print(f"Accuracy of this model is {accuracy}%")
```

```python
def predict_disease(object):
    variable=model.predict(object)[0]
    return disease[variable]
```

```python
def create_input_to_model(symptoms_list, input_symptoms):
    input_to_model = np.zeros(len(symptoms_list), dtype=int)
    for symptom in input_symptoms:
        if symptom in symptoms_list:
            index = symptoms_list.index(symptom)
            input_to_model[index] = 1
    return input_to_model.reshape(1,-1)

# Example usage
symptoms_list = ['cough', 'fever', 'headache', 'nausea', 'fatigue', 'muscle pain', 'sore throat']
input_symptoms = ['fever', 'fatigue', 'sore throat']
input_to_model = create_input_to_model(symptoms_list, input_symptoms)
print(input_to_model)
```

```python
def predict_disease_from_symptoms(symptoms,given_symptom_list):
    input_to_model = create_input_to_model(symptoms,given_symptom_list)
    disease_index=model.predict(input_to_model)[0]
    return disease[disease_index]
```
